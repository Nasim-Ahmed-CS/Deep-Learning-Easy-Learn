```python
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import torch.optim
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = [elem[0] for elem in iris['data']]
X = np.array(X)
X = X.reshape(X.shape[0], 1)

Y = [elem[2] for elem in iris['data']]

Y = np.array(Y)
Y = Y.reshape(Y.shape[0], 1)

XX = X
YY = Y

inputs = torch.from_numpy(X.astype(np.float32))
targets = torch.from_numpy(Y.astype(np.float32))

model = nn.Linear(1, 1)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
inputs/5
```

```python
epochs = 1000
temp = inputs
tempZ = inputs
for i in range(epochs):
  optimizer.zero_grad()
  outputs = model(inputs)
  if i == 0:
    tempZ = outputs
  temp = outputs
  loss = criterion(outputs, targets)
  print(f'Epoch: {i + 1}, Loss: {loss.item()}')
  loss.backward()
  optimizer.step()
```

```python
inputsNP = inputs.numpy()
tempNP = temp.detach().numpy()
tempZNP = tempZ.detach().numpy()
plt.scatter(XX, YY)
plt.plot(inputsNP, tempNP)
plt.plot(inputsNP, tempZNP)
```
